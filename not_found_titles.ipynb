```python
import re 
import pandas as pd
import pickle
import numpy as np
```

```python
jerome_seznam = 'data/jerome_seznam.xlsx'
jerome_found = 'found_titles_synv9_id.csv'
jerome_not_found = "data/not_found_synv9_id.xlsx"
jerome_found_excel = "data/found_synv9_id.xlsx"
```

```python
df_jerome_all = pd.read_excel(jerome_seznam,header=0)
df_jerome_found = pd.read_csv(jerome_found,header=0)

# Merge the dataframes
merged_df = pd.merge(df_jerome_all, df_jerome_found, on='id', how='left', indicator=True)

# Filter out the rows where the title is present in both dataframes
non_found_titles_df = merged_df[merged_df['_merge'] == 'left_only']

# Drop the indicator column '_merge'
non_found_titles_df = non_found_titles_df.drop(columns='_merge')

non_found_titles_df.to_excel(jerome_not_found)   
```

```python
txtype_old = df_jerome_found.txtype.unique()
txtype_new = df_jerome_found.txtype_syn.unique()
for txtype in txtype_old:
    print("Type {txtype}: {count}".format( txtype = txtype, count = sum(df_jerome_found['txtype'] == txtype))  )

print()
for txtype in txtype_new:
    print("Type {txtype}: {count}".format( txtype = txtype, count = sum(df_jerome_found['txtype_syn'] == txtype))  )

print("Type {txtype}: {count}".format( txtype = 'NaN', count = sum(df_jerome_found['txtype_syn'] == 'NaN'))) 
```

```python
df_jerome_not_found = pd.read_excel(jerome_not_found,header=0)

# Merge the dataframes
merged_df = pd.merge(df_jerome_all, df_jerome_not_found, on='id', how='left', indicator=True)

# Filter out the rows where the title is present in both dataframes
found_titles_df = merged_df[merged_df['_merge'] == 'left_only']

# Drop the indicator column '_merge'
found_titles_df = found_titles_df.drop(columns='_merge')

found_titles_df.to_excel(jerome_found_excel)   
```
